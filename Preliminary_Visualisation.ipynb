```python
import h5py
import numpy as np

# Load dataset
with h5py.File("lotka_volterra_data.h5", "r") as f:
    trajectories = f["trajectories"][:]  # Shape: (1000, 100, 2)
    time_points = f["time"][:]           # Shape: (100,)

# Select the first system (system_id = 0) and first 10 time steps
prey = trajectories[0, :10, 0]  # First 10 prey values
predator = trajectories[0, :10, 1]  # First 10 predator values
times = time_points[:10]  # First 10 time steps

print("Time:", times)
print("Prey Population:", prey)
print("Predator Population:", predator)
```

```python
import matplotlib.pyplot as plt

# Select first system
system_id = 0
prey = trajectories[system_id, :, 0]
predator = trajectories[system_id, :, 1]

# Plot the prey and predator populations over time
plt.figure(figsize=(10, 5))
plt.plot(time_points, prey, label="Prey Population", color="blue")
plt.plot(time_points, predator, label="Predator Population", color="red")
plt.xlabel("Time")
plt.ylabel("Population")
plt.title("Lotka-Volterra Predator-Prey Dynamics")
plt.legend()
plt.grid()
plt.show()
```

```python
for system_id in range(3):  # Show first 3 systems
    prey = trajectories[system_id, :, 0]
    predator = trajectories[system_id, :, 1]

    plt.plot(time_points, prey, label=f"Prey System {system_id}")
    plt.plot(time_points, predator, label=f"Predator System {system_id}")

plt.xlabel("Time")
plt.ylabel("Population")
plt.title("Multiple Predator-Prey Systems")
plt.legend()
plt.savefig("Predator_Prey_Systems")
plt.show()
```
